```python
class Item:
    def __init__(self, nombre, precio, cantidad):
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad

    def get_precio(self):
        return self.precio

    def set_precio(self, precio):
        self.precio = precio

    def get_cantidad(self):
        return self.cantidad

    def set_cantidad(self, cantidad):
        self.cantidad = cantidad


class Carrito:
    def __init__(self):
        self.items = []

    def agregar_item(self, item):
        self.items.append(item)

    def quitar_item(self, nombre):
        for item in self.items:
            if item.nombre == nombre:
                self.items.remove(item)
                break

    def contar_num_elementos(self):
        return len(self.items)

    def calcular_precio_total(self):
        total = sum(item.precio * item.cantidad for item in self.items)
        return total

    def limpiar_carrito(self):
        self.items.clear()


def main():
    carrito_compras = Carrito()

    while True:
        print("\nBienvenido a la Tienda en Línea:")
        print("1. Añadir al carrito")
        print("2. Eliminar del carrito")
        print("3. Contar número total de elementos")
        print("4. Calcular precio total de la compra")
        print("5. Limpiar el carrito")
        print("6. Terminar el programa")

        opcion = input("Selecciona una opción (1/2/3/4/5/6): ")

        if opcion == "1":
            nombre = input("Ingrese el nombre del producto: ")
            precio = float(input("Ingrese el precio del producto: "))
            cantidad = int(input("Ingrese la cantidad: "))
            item = Item(nombre, precio, cantidad)
            carrito_compras.agregar_item(item)
            print(f"{cantidad} {nombre}(s) se ha(n) añadido al carrito.")
        elif opcion == "2":
            nombre = input("Ingrese el nombre del producto a eliminar: ")
            carrito_compras.quitar_item(nombre)
            print(f"{nombre} se ha eliminado del carrito.")
        elif opcion == "3":
            print(f"Total de elementos en el carrito: {carrito_compras.contar_num_elementos()}")
        elif opcion == "4":
            print(f"Precio total de la compra: ${carrito_compras.calcular_precio_total():.2f}")
        elif opcion == "5":
            carrito_compras.limpiar_carrito()
            print("El carrito se ha limpiado.")
        elif opcion == "6":
            print("Gracias por comprar con nosotros. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción válida.")


if __name__ == "__main__":
    main()
```
